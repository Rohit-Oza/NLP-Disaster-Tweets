```python
unique_keyword = train.keyword.unique().tolist()
unique_keyword.remove(unique_keyword[0])
unique_keyword
```

```python
unique_keyword_df = pd.DataFrame(unique_keyword)
```

```python
a = Non_Key_List['text'].tolist()
a
```

```python
for i, j in enumerate(a):
    for c in unique_keyword:
        if c in j:
            Non_Key_List.iloc[i,1] = c
```
